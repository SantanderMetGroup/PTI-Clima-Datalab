### 3. Carga de datos y visualización

El inventario nos indica cuál es el `endpoint` de cada dataset. Es decir, el enlace que necesitamos utilizar para la carga de datos.

Primero concretamos el dataset que queremos analizar y extraemos el elemento *endpoint*. Empezaremos por la Península Ibérica.

```r
dataset <- subset(df, dataset == "CHELSA-W5E5v1.0_Iberia_day")
```

```r
endpoint <- dataset$endpoint
```

Antes de cargar los datos, conviene revisar el dataset con `dataInventory`. Esta función devuelve un resumen útil para entender qué información contiene y cómo está descrita (variables, unidades y cobertura temporal, entre otros aspectos).

```r
di <- dataInventory(endpoint)
```

```r
str(di)
```

Este resumen facilita definir la petición de carga de datos. En este ejemplo, cargamos la precipitación diaria (`pr`). La carga se realiza con la función `loadGridData`, que permite especificar otros parámetros además de la variable y el periodo; por ejemplo, podemos seleccionar una estación concreta o agregar los datos a escala mensual. 

Como se observa en el resumen del dataset, la precipitación se expresa en unidades de flujo (kg m-2 s-1). Para obtenerla en milímetros (mm) podemos realizar una transformación posterior a la carga o, alternativamente, utilizar un diccionario de armonización (en este caso, Harmonization_dictionaries/CHELSA-W5E5v1.0.dic) para que los datos se devuelvan en las unidades que climate4R define como estándar. Para más detalles sobre el uso de diccionarios en la armonización de datos, consulta el notebook de primeros pasos (primeros_pasos_R.ipynb).

**La operación de carga es costosa en el caso del dataset CHELSA-W5E5v1.0**, ya que ofrece una resolución de 1 km, lo que implica un gran volumen de datos para la Península Ibérica. Por ello, del periodo completo disponible (véase `Date_range` en el resumen del dataset), solicitaremos un único mes: enero del año 2000.

```r
pr <- loadGridData(dataset = endpoint,
                           var = "pr",
                           season = 1,
                           years = 2000,
                           aggr.m = "mean",
                           dictionary = "../../Harmonization_dictionaries/CHELSA-W5E5v1.0.dic")
```

Las principales funciones de visualización son `temporalPlot` (series temporales) y `spatialPlot` (mapas). Dado que hemos cargado **un único mes** y hemos aplicado la **agregación mensual** durante la carga, no disponemos de datos suficientes para construir una serie temporal. Por lo tanto, generaremos un **mapa** con `spatialPlot`, que nos dará una idea del **dominio** y de la **heterogeneidad espacial**.

```r
spatialPlot(pr, 
            at = seq(0, 3, 0.1),
            set.max = 3,
            set.min = 0,
            main= "Promedio de la precipitación diaria en enero del 2000", 
            backdrop.theme = "coastline", 
            color.theme = "BrBG", 
            colorkey = list(space = "right",
                            title = list("mm/día", cex = 1)
                           ),
            xlab = "Longitud",
            ylab = "Latitud"
           )
```

Para trabajar con estos datos, se recomienda desarrollar **scripts avanzados** capaces de gestionar la **carga** y el **postprocesado** de forma eficiente; por ejemplo:

* descargando únicamente la **zona de interés**,
* operando mediante **bucles espaciales** y/o **temporales**, e
* incorporando las **transformaciones** dentro del propio **bucle de carga**,

con el fin de **evitar la creación de objetos voluminosos**.

A continuación se muestra un **ejemplo básico** en el que realizamos la carga para la región de **Galicia** en un bucle por **años**, calculando simultáneamente la **media estacional**. El resultado será una **lista de *grids*** (un *grid* por año) que posteriormente uniremos en un único *grid* mediante `bindGrid`.


```r
years <- 1991:2010

pr.by.years <- lapply(years, function(y) {
    message("Loading year ", y)
    pr <- loadGridData(dataset = endpoint,
                           var = "pr",
                           season = c(12, 1:2),
                           lonLim = c(-9.5, -6.5),
                           latLim = c(41.7, 43.8),
                           years = y,
                           aggr.m = "mean",
                           dictionary = "../../Harmonization_dictionaries/CHELSA-W5E5v1.0.dic") %>% suppressMessages()
    climatology(pr) %>% suppressMessages()
})
```

```r
pr.annual <- bindGrid(pr.by.years, dimension = "time")
```

El objeto resultante es lo que denominamos *grid de climate4R*. Se trata de una lista con los elementos "Variable", "Data", "xyCoords" y "Dates". Es posible extraer cada uno de estos componentes y transformarlos mediante operaciones básicas de R, sin embargo, las librerías de climate4R están diseñadas para trabajar de forma transparente con la estructura completa.

```r
str(pr.annual)
```

*climate4R* incluye múltiples funciones de transformación para agregar datos (temporal y espacialmente), extraer subconjuntos, modificar unidades, etc. A continuación se muestran algunos ejemplos. 

Para obtener el promedio temporal a lo largo de todo el periodo del grid, la función más directa es `climatology`.

```r
pr.clim <- climatology(pr.annual)
```

Para extraer localidades o subconjuntos espaciales usaremos `subsetGrid` definiendo los argumentos `lonLim` y `latLim`. Para extraer una subregión, `latLim` y `lonLim` se definen como vectores de dos elementos que definen los límites de la región de interés. En este ejemplo, extraemos la información para localidades puntuales, por lo tanto se asigna un único valor a `latLim` y a `lonLim`. 

```r
lugo <- subsetGrid(pr.annual, latLim = 43, lonLim = -7.56)
carballo <- subsetGrid(pr.annual, latLim = 43.21, lonLim = -8.69) 
verin <- subsetGrid(pr.annual, latLim = 41.94, lonLim = -7.44)
```

Las principales funciones de visualización son `temporalPlot` (series temporales) y `spatialPlot` (mapas).

```r
temporalPlot("Lugo" = lugo, "Carballo" = carballo, "Verín" = verin, 
             xyplot.custom = list(main = "Media invernal de la precipitación diaria (mm/día)"))
```

```r
spatialPlot(pr.clim, 
            at = seq(0, 7, 0.2),
            set.max = 7,
            set.min = 0,
            main= "Promedio invernal de la precipitación diaria en el periodo 1991-2010", 
            color.theme = "BrBG", 
            colorkey = list(space = "right",
                            title = list("mm/día", cex = 1)
                           ),
            xlab = "Longitud",
            ylab = "Latitud"
           )
```

En lugar de la media podemos calcular cualquier otro estadístico con `climatology` definiendo la función de agregación (parámetro `clim.fun`), por ejemplo, podemos visualizar el percentil 90 utilizando previamente la función `quantile` (librería `stats` de R) en el cálculo de la climatología.

A continuación se repite parte del proceso visto anteriormente, considerando el dataset correspondiente a ***Canarias***. En este caso cargaremos la temperatura máxima de verano. Primero cargamos y transformamos los datos.

```r
dataset.can <- subset(df, dataset == "CHELSA-W5E5v1.0_Canarias_day")
endpoint.can <- dataset.can$endpoint

years <- 1991:2010
tasmax.can.by.years <- lapply(years, function(y) {
    message("Loading year ", y)
    tx <- loadGridData(dataset = endpoint.can,
                           var = "tasmax",
                           season = 6:8,
                           years = y,
                           aggr.m = "mean",
                           dictionary = "../../Harmonization_dictionaries/CHELSA-W5E5v1.0.dic") %>% suppressMessages()
    climatology(tx) %>% suppressMessages()
})

tasmax.can.anual <- bindGrid(tasmax.can.by.years, dimension = "time")
tasmax.can.clim <- climatology(tasmax.can.anual)
```

A continuación visualizamos los resultados. En este ejemplo, en lugar de extraer localidades concretas de las islas, representaremos la serie correspondiente a la media regional de todas las islas. La función `temporalPlot` realiza esta aggregación regional de manera automática. 

```r
temporalPlot("Canarias" = tasmax.can.anual)

spatialPlot(tasmax.can.clim, 
            at = seq(22, 34, 1),
            set.max = 34,
            set.min = 22,
            main= "Promedio de la temperatura máxima de verano en el periodo 1991-2010", 
            color.theme = "YlOrRd", 
            colorkey = list(space = "right",
                            title = list("º C", cex = 1)
                           ),
            xlab = "Longitud",
            ylab = "Latitud"
           )
```
